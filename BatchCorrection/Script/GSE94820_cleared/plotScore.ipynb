```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = 'E:/2021-spring/batch/IntegratedCode_batchCorrection/Dataset/GSE94820_cleared/score/'

raw_lisi = pd.read_csv(path+'raw_lisiScore.csv')
BBKNN_lisi = pd.read_csv(path+'BBKNN_lisiScore.csv')
COMBAT_lisi = pd.read_csv(path+'COMBAT_lisiScore.csv')
harmony_lisi = pd.read_csv(path+'harmony_lisiScore.csv')
iMAP_lisi = pd.read_csv(path+'iMAP_lisiScore.csv')
RISC_lisi = pd.read_csv(path+'RISC_lisiScore.csv')
```

```python
datalist = [raw_lisi,BBKNN_lisi,COMBAT_lisi,harmony_lisi,iMAP_lisi,RISC_lisi]
namelist = ['raw','BBKNN','combat','Harmony','iMAP','RISC']

for i in range(6):
    plt.scatter(np.mean(datalist[i])[0],np.mean(datalist[i])[1],label = namelist[i])

plt.xlabel('iLISI')
plt.xlim(1,2)
plt.ylabel('cLISI')
plt.grid()
plt.legend()
plt.savefig('figures/lisi.svg')
plt.show()
```
