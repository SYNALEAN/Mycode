```python
import pandas as pd
import matplotlib.pyplot as plt

species = pd.read_csv('E:\\Jupyter\\singleSpecies.csv',header=0)
type1 = species['Climate'].tolist()

qaqq = {'tropical rain forests':1,'arboreal': 2,
                 'temperate':3,'semi-arid':4,'arid':5}
species['Climate_iden']=species['Climate'].map(qaqq)
species
```

```python
name = species['name3'].value_counts()
name = name.index
name = name.tolist()
```

```python
plt.figure(figsize=(10,8),facecolor='white')
#plt.subplot2grid((3, 2), (0, 0), colspan=2, rowspan=2)
plt.grid()
for item in name:
    
    parttable = species[species['name3']==item]
    plt.scatter(parttable['Climate_iden'],parttable['Means'],label = item)

plt.ylabel('Average Decomposition Rate (55d\'s growth)')
plt.xlim((0,6))
plt.xticks( range(1,6), ('Tropical rain forests','Arboreal', 'Temperate', 'Semi-arid', 'Arid') )
plt.legend(ncol = 3,bbox_to_anchor=(1.14,1))


plt.savefig('s_sto_ps.jpg')
```
