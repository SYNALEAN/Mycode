```python
import os
path = 'E:/课题组相关/科创II/RNA-seq-calculated/'
files_= os.listdir(path)
```

```python
import  pandas as pd
files_= os.listdir(path)
expressionInAll = pd.read_csv(path+files_[0], compression='gzip',header=0)
expressionInAll.rename(columns={'TPM': files_[0]}, inplace=True)

del files_[0]
```

```python
for item in files_:
    DF = pd.read_csv(path+item, compression='gzip',header=0)
    DF.rename(columns={'TPM': item}, inplace=True)
    expressionInAll = pd.merge(expressionInAll, DF, how='left', on = ['loadedanno.gene_id'])
    print('finished_'+item)
```
