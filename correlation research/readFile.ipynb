```python
import os
import gzip
import pandas as pd


path = "E:/课题组相关/科创II/RNA-seq/"
files= os.listdir(path)
path2 = 'E:/课题组相关/科创II/RNA-seq-calculated/'
Exonanno = pd.read_csv('geneid_anno_1.csv',header=0)
Exonanno.rename(columns={'1': 'exonlen'}, inplace=True)
def getfilename(filepath):
    fileName = os.listdir(filepath)
    maxlength = 0
    maxindex = 0
    for index in range(0,len(fileName)):
        thisname = fileName[index]
        if maxlength <= len(thisname):
            maxlength = len(thisname)
            maxindex = index
    
    return fileName[maxindex]
```

```python
for i in range(0,len(files)):
    test = files[i]
    testpath = path + test
    fileName = getfilename(testpath)
    DF = pd.read_csv(testpath+'/'+fileName, compression='gzip',sep='\t',header=None)
    DF.rename(columns={0: 'loadedanno.gene_id'}, inplace=True)
    merged = pd.merge(DF,Exonanno[['loadedanno.gene_id','exonlen']])
    merged.dropna(axis=0,how='any')
    rperL = merged[1]/merged['exonlen'] #这里可能出错？
    fenmu = rperL.sum()
    tpm = rperL/fenmu*1000000
    resultedDF = pd.concat([merged['loadedanno.gene_id'], tpm], axis=1)
    resultedDF.rename(columns={0: 'TPM'}, inplace=True)
    resultedDF.to_csv(path2+fileName,index=0) #虽然叫这个名字但是确实是csv格式？
    print(fileName + '_finished')
    print(str(i))
```
